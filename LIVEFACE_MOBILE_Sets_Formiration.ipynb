```python
import os

from os import listdir
from collections import Counter
from os.path import join, isfile
```

# Collect Data

```python
top = 'LIVEFACE_DATA/'
```

```python
directories = {}
```

```python
counter = 0
for root, dirs, _ in os.walk(top, topdown = False):
    for name in dirs:
        if len(listdir(join(root, name))) > 20:
            path_dir = join(root,name)
            
#             p
            
            label = 'LIVE' if path_dir.rfind('LIVE') > 0 else 'FAKE'
            files = [join(path_dir,file) for file in listdir(path_dir)]
            
            
            directories[path_dir] = {
                'label' : label,
                'files' : files,
                'count_of_images' : len(files)
            }
```

```python
labels = []
file_names = []
directories_names = []


for key in directories:
    
    sample = directories[key]
    
    file_names.extend(sample['files'])
    directories_names.extend([key] * sample['count_of_images'])
    labels.extend([sample['label']] * sample['count_of_images'])
```

```python
dataset = {
    'image_path' : file_names,
    'image_metadata' : directories_names,
    'label' : labels
}
```

```python
dataset = pd.DataFrame(dataset)
```

```python
dataset.to_csv('LIVEFACE_Data_Representation.csv',index = False)
```
